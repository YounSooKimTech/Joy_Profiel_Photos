# Example Code for Face and Smile Recognition with OpenCV Library

## Dataset: http://vis-www.cs.umass.edu/lfw/#download
## OpenCV smile detection: http://visilab.etsii.uclm.es


```python
import cv2
from google.colab.patches import cv2_imshow

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')

img = cv2.imread("/content/George_W_Bush_0322.jpg", -1)

faces = face_cascade.detectMultiScale(img, scaleFactor=1.3, minNeighbors=5)

for (x, y, w, h) in faces:
    # Draw rectangle around face
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_bottom_half = img[y + int(h * 0.6):y + h, x:x + w]

    if roi_bottom_half.shape[0] > 0 and roi_bottom_half.shape[1] > 0:  # Check if the region of interest is not empty
        smiles = smile_cascade.detectMultiScale(roi_bottom_half, scaleFactor=1.3, minNeighbors=30)
        for (min_x, min_y, max_w, max_h) in smiles:
            cv2.rectangle(img, (x + min_x, y + int(h * 0.6) + min_y), (x + min_x + max_w, y + int(h * 0.6) + min_y + max_h), (0, 255, 0), 1)

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
